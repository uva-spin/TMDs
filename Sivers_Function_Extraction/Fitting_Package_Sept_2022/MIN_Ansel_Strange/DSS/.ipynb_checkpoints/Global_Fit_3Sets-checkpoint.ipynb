```python
def totalchi2Minuit(m1,Nu,au,bu,Nub,Nd,ad,bd,Ndb,Ns,aS,bS,Nsb):
    tempchi2=SIDIStotalchi2Minuit(m1=m1,Nu=Nu,au=au,bu=bu,Nub=Nub,Nd=Nd,ad=ad,bd=bd,
                                  Ndb=Ndb,Ns=Ns,aS=aS,bS=bS,Nsb=Nsb)
    return tempchi2
```
